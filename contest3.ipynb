```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

import warnings
warnings.filterwarnings('ignore')
```

```python
fig, ax=plt.subplots(figsize=(15,15))
corr= tdf4[['online', 'offline', 'internship', 'very bad', 'bad', 'normal', 'good', 'very good']].corr()

mask= np.zeros_like(corr, dtype=bool)
mask[np.triu_indices_from(mask)]=True
sns.heatmap(
    corr,
    mask=mask,
    fmt='.2f',
    annot=True,
    ax=ax,
    cmap='RdBu',
    vmax=1,
    vmin=-1,
    square=True
)
```

## **df1에 대해서 위와 같이 피어슨 상관계수 작업 실행**
**교육 이수자와 개선점 간의 관계**

```python
df1.drop("Ind",axis=1,inplace=True)
tdf1=df1.T
tdf1.columns
```

```python
# 컬럼이름 바꾸기
tdf1.columns=['online', 'offline', 'internship', 'contents', 'level', 'professional', 'cost support', 'field training','time diversity','geo_acessibility','oe_expansion']
tdf1
```

```python
c_tdf1=tdf1[['online', 'offline', 'internship', 'contents', 'level', 'professional', 'cost support', 'field training','time diversity','geo_acessibility','oe_expansion']].corr(method='pearson')
c_tdf1
```

```python
fig, ax=plt.subplots(figsize=(15,15))
corr= tdf1[['online', 'offline', 'internship', 'contents', 'level', 'professional', 'cost support', 'field training','time diversity','geo_acessibility','oe_expansion']].corr(method='pearson')

mask= np.zeros_like(corr, dtype=bool)
mask[np.triu_indices_from(mask)]=True
sns.heatmap(
    corr,
    mask=mask,
    fmt='.2f',
    annot=True,
    ax=ax,
    cmap='RdBu',
    vmax=1,
    vmin=-1,
    square=True
)
```
