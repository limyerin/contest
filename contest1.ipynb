100이상이면 사례수(명)으로 파악해서 백분율(100이하) 와 곱해주는 반복문 만들려고 시도.
반복문 진행하다가 다시 100이상의 수를 만나면 다시 곱해서 구하는 방식 생각.

```python
for i in range(1,51):
  if (data1[i-1:i]['y19']>=100):
    j=i
    n=0
    while((data1[j:j+1]['y19']<100) == True) :
      n+=1
      re= data1[i-1:i]['y19']*(data1[j:j+1]['y19'])
      data1.replace(data1[j:j+1]['y19'], re)
      j+=1
    
    i+=n

  if (data1[i-1:i]['y19'] == np.NaN):
      pass

  else:
      pass



```
